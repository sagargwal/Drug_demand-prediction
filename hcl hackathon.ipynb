```python
# REMOVING OUTLIER IN DEMAND
# using iqr method 
# removing outlier
q1 = data.Demand.quantile(0.25)
q3 = data.Demand.quantile(0.75)
iqr = q3 - q1
u_outlier = q3+1.5*iqr
l_outlier = q1-1.5*iqr
data = data[(l_outlier<=data.Demand) & (data.Demand<=u_outlier)]
```

```python
# the data is normal and does not follow any trend lets confirm using skewness
from scipy.stats import skew

skewness = data.Demand.skew()
print(f"Skewness: {skewness}")
```

```python
import numpy as np 
n = len(data.Demand)
SES = np.sqrt(6 / len(data))

skew_ratio = abs(skewness / SES)
skew_ratio
```

```python
# the drug type demand distribution does not look significant
# calculating variance
print(data.Demand[data.Drug_Type == "Oral"].var(),data.Demand[data.Drug_Type == "Injectable"].var(),data.Demand[data.Drug_Type == "Topical"].var())
```

```python
# varince look similar anova can be done
from scipy.stats import f_oneway
f_oneway(data.Demand[data.Drug_Type == "Oral"],data.Demand[data.Drug_Type == "Injectable"],data.Demand[data.Drug_Type == "Topical"])
```

```python
# filling na in main
plt.figure(figsize=(8,6))
sns.boxplot(x="Main", y="Demand", data=data)
plt.title("Distribution of Drug Demand by Main")
plt.xticks(rotation=45)  # Rotate labels if needed
plt.grid(True)
plt.show()
```

```python
# variance for diferent category of main
main_unique = data.Main.unique()
main_dict = {}
for i in main_unique:
    var = data[data.Main==i].Demand.var()
    main_dict[i] = var
main_dict
```

```python
# the variance of each group is different so we can not perform parametric test to know the influence of sides on demand
import scipy.stats as stats
main_value_dict = {}
for i in main_unique:
    group = data[data.Main==i].Demand
    main_value_dict[i] = group
stats.kruskal(main_value_dict['S101'],main_value_dict['S102'],main_value_dict['S!03'],main_value_dict['Nausea, Headache'],main_value_dict['Fatigue, Vomiting'])
```

```python
# p value is less then 0.05 so the data here is significant and we will keep this 
data.Main.fillna("unknown",inplace = True)
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
from sklearn.impute import SimpleImputer

# Handle missing values if any
imputer = SimpleImputer(strategy="mean")
x_train = imputer.fit_transform(x_train)
x_test = imputer.transform(x_test)

# Initialize XGBoost model
reg1 = xgb.XGBRegressor(
    n_estimators=1000,
    early_stopping_rounds=50,
    learning_rate=0.01
)

# Train model
reg1.fit(
    x_train, y_train,
    eval_set=[(x_train, y_train), (x_test, y_test)],
    verbose=50
)

# Make predictions
y_pred = reg1.predict(x_test)

# Evaluate
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
print(f"root mean squared value: {np.sqrt(mse)}" )
```

```python
ax = y_test.plot(figsize=(15, 5))
y_pred.plot(ax=ax, style="-")
plt.legend(['Truth Data', 'Predictions'])
ax.set_title('Raw Dat and Prediction')
plt.show()
```
